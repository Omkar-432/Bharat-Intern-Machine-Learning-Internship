# Iris Flower Classfication using SVC, Decision Tree Classifier and Random Forest

Importing Dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

Loading Dataset

```python
Dataset = pd.read_csv('Iris Flower Data.csv')
```

```python
Dataset.head()
```

# Data Pre-Processing and Exploration

Dropping 'Id' attribute

```python
Dataset = Dataset.drop(labels='Id', axis=1)
```

```python
Dataset.describe()
```

```python
Dataset['Species']
```

```python
Dataset['Species'].value_counts()
```

```python
ax = plt.subplots(1, 1, figsize=(10,5))
sns.countplot(x=Dataset['Species'])
plt.title("Iris Species Count")
plt.show()
```

```python
Dataset.boxplot(by = 'Species', figsize=(12,10))
plt.show
```

```python
sns.pairplot(Dataset, hue='Species')
```

```python
print(Dataset.corr)
```

```python
sns.heatmap(Dataset.corr(numeric_only = True), cmap="YlGnBu", annot=True)
plt.show()
```

Seperating Features and Target variable

```python
Datanew  = Dataset.values
X = Datanew[:,0:4]
Y = Datanew[:,4]
```

```python
print(X)
```

```python
print(Y)
```

```python
X.shape
```

```python
Y.shape
```

Using Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# Model 1 - Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python
prediction1 = model_svc.predict(X_test)
```

```python
prediction1
```

Printing Actual value against Predicted value

```python
for i in range(len(prediction1)):
    print(Y_test[i], prediction1[i])
```

Printing Accuracy Score

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1))
```

# Model 2 - Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, Y_train)
```

Printing Accuracy Score

```python
prediction2 = model_DTC.predict(X_test)
print(accuracy_score(Y_test, prediction2))
```

Printing Actual value against Predicted value

```python
for i in range(len(prediction2)):
    print(Y_test[i], prediction2[i])
```

# Model 3 - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier()
```

```python
model_rf.fit(X_train, Y_train)
```

```python
prediction3 = model_rf.predict(X_test)
```

```python
print(model_rf.score(X_test, Y_test))
```

```python
for i in range(len(prediction3)):
    print(Y_test[i], prediction3[i])
```
